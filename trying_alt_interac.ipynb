```python
import pandas as pd
import altair as alt
from ipywidgets import interact
```

```python
# Data Loading 
attention = pd.read_stata("../HW/static_project/attended_09_19.dta")
attention_peru = attention.loc[attention['dpto'] == 0]
attention_costa = attention.loc[attention['region'] == 'Costa']
attention_sierra = attention.loc[attention['region'] == 'Sierra']
attention_selva = attention.loc[attention['region'] == 'Selva']
```

```python
attention_costa
```

```python
# Chart 2 
chart2=alt.layer(
    alt.Chart(attention_peru).mark_bar(color='#eee').encode(alt.X("total_dist:Q", scale=alt.Scale(nice=False), title=None)),
    alt.Chart(attention_peru).mark_bar().encode(color=alt.Color('measures[0]:N', scale=alt.Scale(domain=["2009", "2019", "Targeted districts"], range=["#d62728", "#ff9896", "black"]))),
    alt.Chart(attention_peru).mark_bar(color='#ff9896', size=10).encode(x='upto_2019:Q', ),
        alt.Chart(attention_peru).mark_bar(color='#d62728', size=10).encode(x='upto2009:Q', ),
    alt.Chart(attention_peru).mark_tick(color='black', thickness=2).encode(x='targeted:Q'),
    width=350,
    height=30).facet(
        title={"text": ["Regions differ vastly in total and targeted districts", "Expansion of CunaMas has been heterogeneous"], 
                        "subtitle": ["", "Number of attended districts by region", ""],
                        "color": "black", "subtitleColor": "black", "align":"center", 
                        "font": "LM Roman", "fontSize":20, "fontWeight": "bold",}, 
        row=alt.Row("dpto_str:O", sort=['Peru'], title=None)
        ).resolve_scale(x='independent')

chart2 = alt.concat(chart2,
    title=alt.TitleParams(
        'Source: CunaMas Administrative Records',
        baseline='bottom',
        orient='none',
        anchor='end',
        dx=375, dy=105,
        fontSize=10))
                      
chart2.configure_legend(title=None, direction='horizontal', orient="none", 
                      legendX=75, legendY=65, padding=5, rowPadding=5, 
                      strokeColor='Gray').configure_title(
                          fontSize=20, offset=5, orient='top', anchor='middle')
```

## This is drop down graphic

```python
ex = pd.read_csv(r'./data/example1.csv', parse_dates = ['year'], date_parser=lambda x:pd.to_datetime(x, format='%Y'))
ex['year'] = ex['year'].dt.strftime('%Y')
ex
```

```python
my_cmap = {
    'Not targeted': '#fdc1c5',
    'Attended': '#fc824a',
    'Target not attended': '#9f2305',
}

title = ["This is the title"]
subtitle = ["and this the subtitle"] 

dropdown_indicator = alt.binding_select(name='Indicator: ', options=[None]+list(ex["outcome"].unique())
                                       ,labels=['All Indicators']+ list(ex["outcome"].unique()))
dropdown_comparison = alt.binding_select(name='Zoom Group: ', options=[None]+list(ex["group"].unique())
                                         ,labels=['All Groups']+ list(ex["group"].unique()))
#year_slider = alt.binding_range(min=2008, max=2019, step=1)

selection_indicator = alt.selection_single(fields=["outcome"], bind=dropdown_indicator)
selection_comparison = alt.selection_single(fields=["group"], bind=dropdown_comparison)
#slider_selection = alt.selection_single(bind=year_slider, fields=['year'], name="year")

chart = alt.Chart(ex).mark_line().encode(
    x=alt.X('year:T', title="Year", scale=alt.Scale(zero=False), 
            axis=alt.Axis(grid=False)), 
    y=alt.Y('value', title="Share in %", scale=alt.Scale(zero=False)),
    color=alt.Color('group', title='',
                    scale=alt.Scale(
                        domain=list(my_cmap.keys()), 
                        range=list(my_cmap.values())),
                    legend=alt.Legend(
                        title="Districts", 
                        #orient='top-left',
                        #direction = 'horizontal',
                        legendX=235, legendY=151, padding=2, 
                        rowPadding=5, strokeColor='Gray')),
    tooltip= [alt.Tooltip('group:N'),
               alt.Tooltip('value:Q'),
               alt.Tooltip('year:T')
              ])


chart = chart.add_selection(selection_comparison, selection_indicator, #slider_selection,
            ).transform_filter(selection_comparison
            ).transform_filter(selection_indicator
            )#.transform_filter(slider_selection)

chart = chart.properties(title={ "text": title,  "subtitle": subtitle,
                        "color": "black", "subtitleColor": "black", "align":"center", 
                        "font": "LM Roman", "fontSize":20, "fontWeight": "bold",},
                        width=400, height=300)


chart = alt.concat(chart,
    title=alt.TitleParams(
        'Source: DHS 2009-2019',
        baseline='bottom',
        orient='none',
        dx=400, dy=350,
        anchor='end',
        fontSize=10))

chart.properties(
    width='container',
    height='container')
```
